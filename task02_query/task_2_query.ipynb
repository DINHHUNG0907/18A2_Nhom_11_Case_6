2. Đọc dữ liệu sạch

```python
import pandas as pd

df_candidate = pd.read_csv("data_clean/candidate_profile_clean.csv")
df_interview = pd.read_csv("data_clean/interview_score_clean.csv")
df_salary = pd.read_csv("data_clean/salary_offer_clean.csv")
```

5. Tính điểm trung bình theo vòng phỏng vấn
Yêu cầu

Tính điểm trung bình theo vòng

```python
avg_score_by_round = (
    df_interview.groupby("round")["score"]
    .mean()
    .reset_index(name="avg_score")
)

avg_score_by_round
```

6. Tìm ứng viên có kinh nghiệm nhưng điểm thấp
Định nghĩa hợp lý (ghi rõ trong notebook)

Kinh nghiệm ≥ 3 năm

Điểm phỏng vấn < 6

```python
# Tính điểm trung bình theo ứng viên
avg_score_candidate = (
    df_interview.groupby("candidate_id")["score"]
    .mean()
    .reset_index(name="avg_score")
)

# Merge với kinh nghiệm
candidate_score = avg_score_candidate.merge(
    df_candidate,
    on="candidate_id",
    how="left"
)

low_score_exp = candidate_score[
    (candidate_score["experience_years"] >= 3) &
    (candidate_score["avg_score"] < 6)
]

low_score_exp
```
